```r
library(dplyr)
library(tidyverse)
# import data
Cleaned_data <- read.csv("Cleaned_data.csv")

# Select the columns we are interested in 
interested_data <- Cleaned_data %>%
  select(Country.x, iso3c, Deaths...million, Pop..density..per.km2..2021.,Sex.ratio..m.per.100.f., Obesity.rate...., Percent)

# change some column names so that they are more readable
interested_data <- interested_data %>%
  rename(Country = Country.x, Deaths.per.million = Deaths...million, population.density = Pop..density..per.km2..2021.,
         Obesity.rate = Obesity.rate...., Vacination.rate = Percent)
# note that vaccination rate can be higher than 100% because it calculates all the vacinated residents, domestic and 
# international alike, divided by the domestic population. These are still valid data and should not be altered.

# since most the missing values are in the same row, meaning we are often missing multiple data from the
# same countries, we might as well drop the countries with missing values in them
interested_data <- na.omit(interested_data)

# Export csv file
#write.csv(interested_data,"interested_data.csv", row.names=FALSE)

# calculate the Pearson correlations between Deaths.per.million and its possible influencial factors
# this is wrote by ChatGPT

# Specify the column to correlate with (Deaths.per.million in this case)
column_to_correlate <- interested_data$Deaths.per.million

# Columns to correlate with (population.density, Sex.ratio, Obesity.rate, Vacination.rate)
columns_to_correlate_with <- interested_data[, c('population.density', 'Sex.ratio..m.per.100.f.', 'Obesity.rate', 'Vacination.rate')]

# Calculate Pearson correlations for each pair
correlations <- sapply(columns_to_correlate_with, function(col) cor(column_to_correlate, col))

# Print the results
result_df <- data.frame(
  Variable = names(correlations),
  Pearson_Correlation = correlations
)

print(result_df)
```

```r
# ok we got positive correlation between death rate and population.density, great. Now let's see if the visualizaton shows the same result 
ggplot(interested_data, aes(x= Deaths.per.million, y = population.density))+
geom_point() +
geom_text(x= max(interested_data$Deaths.per.million),
          y= max(interested_data$population.density),
          label = paste("Correlation =", round(-0.07288994, 4)))+
ggtitle("Correlation between Population Density and Deaths/million")+
geom_smooth() +
theme_minimal()
```

```r
# ok we got positive correlation between death rate and sex.ratio..m.per.100.f., great. Now let's see if the visualizaton shows the same result 
ggplot(interested_data, aes(x= Deaths.per.million, y = Sex.ratio..m.per.100.f.))+
geom_point() +
geom_text(x= max(interested_data$Deaths.per.million),
          y= max(interested_data$Sex.ratio..m.per.100.f.),
          label = paste("Correlation =", round(-0.18734490, 4), hjust = 1, vjust =1))+
ggtitle("Correlation between Sex.ratio..m.per.100.f. and Deaths / million")+
geom_smooth()
```

```r
# ok we got positive correlation between death rate and obesity rate, great. Now let's see if the visualizaton shows the same result 
ggplot(interested_data, aes(x= Deaths.per.million, y = Obesity.rate))+
geom_point() +
geom_text(x= max(interested_data$Deaths.per.million),
          y= max(interested_data$Obesity.rate),
          label = paste("Correlation =", round(0.39567519, 4), hjust = 1, vjust =1))+
ggtitle("Correlation between Obesity rate and Deaths / million")+
theme_minimal() +
geom_smooth()
```
