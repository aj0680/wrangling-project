```r
library(tidyverse)
```

```r
UN_data <- read.csv("final_country_data.csv")
```

Transforming the data into a wider format

```r
UN_wide <- UN_data %>%
  pivot_wider(names_from = Variable, values_from = Value)
```

Selecting a smaller subset of columns.

```r
UN_wide <- UN_wide %>% select(1:30)
UN_wide
```

Cleaning the records - alphabet and spaces are removed.

The string '...' represents missing values in the UN dataset. These values were replaced with NA.

```r
UN_wide_clean <- UN_wide %>%
  mutate(across(c(3:4, 6, 8:9, 11:30),
                ~ na_if(str_replace_all(., "[[:alpha:], ]", ""), "...")))
UN_wide_clean
```

Below cells can be used to randomly sample one record from each dataframe to check the cleanliness of data.

```r
random_index <- sample(1:nrow(UN_wide), 1)
UN_wide[random_index, ]
```

```r
UN_wide_clean[random_index, ]
```

Removing variables with more than 50% missingness and saving the data as a csv

```r
UN_wide_clean <- UN_wide_clean %>%
  select_if(~ mean(is.na(.)) <= 0.50)

write_csv(UN_wide_clean, "UN_countries_clean.csv")
```
